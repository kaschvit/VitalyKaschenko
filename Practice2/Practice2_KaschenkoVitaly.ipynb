##### Задание 13. Заменить пропущенные значения значением среднего в DataFrame

```python
import pandas as pd   
myDF = pd.read_csv('crypto-markets.csv', sep = ';')
for i in range(myDF.shape[1]):
    if (myDF.iloc[:,i].dtypes) == 'float64':
        myDF.iloc[:,i].fillna(myDF.iloc[:,i].mean(axis = 0), inplace = True)
myDF.head()
```

##### Задание 14. Заменить пропущенные значения значением медианны в DataFrame

```python
import pandas as pd   
myDF = pd.read_csv('crypto-markets.csv', sep = ';')
for i in range(myDF.shape[1]):
    if (myDF.iloc[:,i].dtypes) == 'float64':
        myDF.iloc[:,i].fillna(myDF.iloc[:,i].median(axis = 0), inplace = True)
myDF.head(10)
```

##### Задание 15. Заменить пропущенные значения значением максимальным в DataFrame

```python
import pandas as pd   
myDF = pd.read_csv('crypto-markets.csv', sep = ';')
for i in range(myDF.shape[1]):
    if (myDF.iloc[:,i].dtypes) == 'float64':
        myDF.iloc[:,i].fillna(myDF.iloc[:,i].max(axis = 0), inplace = True)
myDF.head(10)
```

##### Задание 16. Заменить пропущенные значения значением минимальным в DataFrame

```python
import pandas as pd   
myDF = pd.read_csv('crypto-markets.csv', sep = ';')
for i in range(myDF.shape[1]):
    if (myDF.iloc[:,i].dtypes) == 'float64':
        myDF.iloc[:,i].fillna(myDF.iloc[:,i].min(axis = 0), inplace = True)
myDF.head(10)
```
